```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from prml.features import GaussianFeatures
```

```python
x = np.linspace(-5, 5, 30)

mean = np.array([0, 2, 4])

feature = GaussianFeatures(mean, 5)

x_ = feature.transform(x)
```

```python
for i in range(0, mean.shape[0]):
    plt.plot(x, x_[:, i], label="mean = %d"%mean[i])
plt.grid()
plt.legend()
plt.show()
```
